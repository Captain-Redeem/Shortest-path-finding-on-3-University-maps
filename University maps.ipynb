#University Maps

##Map 1

image.png

```python
# Define the new edges, using weight=value to assign weights to the edges.
map_1 = [('Bus Stop', 'Library',{'cost':0.1}),
  ('Library', 'Student Centre',{'cost':0.2}), ('Library', 'Car Park',{'cost':1.1}),
  ('Car Park', 'Maths Building',{'cost':0.3}), ('Car Park', 'Store',{'cost':1.2}),
  ('Maths Building', 'Canteen',{'cost':1.3}),
  ('Student Centre', 'Store',{'cost':1.4}),('Student Centre', 'Theatre',{'cost':2.2}),
  ('Store', 'Sports Centre',{'cost':2.3}),('Store','Canteen',{'cost':2.6}),
  ('Canteen','AI Lab',{'cost':2.4}),
  ('Theatre','Sports Centre',{'cost':1.7}),
  ('Sports Centre','AI Lab',{'cost':2.5}),
             ]
```

```python
G = nx.Graph()
```

```python
G.add_edges_from(map_1)
```

```python
nx.draw(G, with_labels=True,node_color='y',node_size=800)
```

##All possible paths

```python
#adding start and end nodes
start_node = 'Bus Stop'
end_node = 'AI Lab'
```

```python
all_paths = list(nx.all_simple_paths(G, start_node, end_node))
for path in all_paths:
    print(path,"\n","Cost:",nx.path_weight(G, path, weight="cost"))
```

###BFS

```python
import networkx as nx
import matplotlib.pyplot as plt

G.add_edges_from(map_1)
# Get the shortest path
shortest_path = nx.shortest_path(G, start_node, end_node)
print(shortest_path)

# Get node positions using spring_layout (or any other layout)
pos = nx.spring_layout(G)

# Draw the graph using nx.draw with the calculated positions
nx.draw(G, pos=pos, with_labels=True, node_color='y', node_size=800)

# Highlight nodes in the shortest path
nx.draw_networkx_nodes(G, pos=pos, nodelist=shortest_path, node_color='r', node_size=800)
plt.show()
```

####BFS Cost

```python
def bfs_cost(graph, start_node, end_node):
    path = shortest_path  # Get the shortest path using BFS
    cost = nx.path_weight(graph, path, weight="cost")  # Path length (number of edges) is the cost
    return cost

cost = bfs_cost(G, start_node, end_node)

print(f"BFS path weight: {cost}")
```

###Bellman-Ford

```python
G.add_edges_from(map_1)
print(nx.shortest_path(G, start_node, end_node, method='bellman-ford'))

# Get node positions using spring_layout (or any other layout)
pos = nx.spring_layout(G)

# Draw the graph using nx.draw with the calculated positions
nx.draw(G, pos=pos, with_labels=True, node_color='y', node_size=800)

# Highlight nodes in the shortest path
nx.draw_networkx_nodes(G, pos=pos, nodelist=shortest_path, node_color='r', node_size=800)

plt.show()
```

####Bellman-Ford cost

```python
#Calculate cost for bellman ford networkx package
def bellman_ford_cost(graph, start_node, end_node):
    path = nx.shortest_path(graph, source=start_node, target=end_node, method='bellman-ford')
    cost = nx.path_weight(graph, path, weight="cost")  # Path length (number of edges) is the cost
    return cost

cost = bellman_ford_cost(G, start_node, end_node)

print(f"Bellman-Ford Path weight: {cost}")
```

###AStar

```python
G.add_edges_from(map_1)
print(nx.astar_path(G, 'Bus Stop', 'AI Lab'))
```

####AStar cost

```python
# Calculate Astar path cost
def astar_cost(graph, start_node, end_node):
    # Define a heuristic function (degree centrality in this case)
    def heuristic(node1, node2):  # Update the heuristic function definition
        return nx.degree_centrality(graph)[node1]

    path = nx.astar_path(graph, source=start_node, target=end_node, heuristic=heuristic, weight='cost')

    cost = nx.path_weight(graph, path, weight="cost")  # Path length (number of edges) is the cost

    return cost

cost = astar_cost(G, start_node, end_node)

print(f"Total cost (A*): {cost}")
```
